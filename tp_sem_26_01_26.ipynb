```python
def FizzBuzz(s : str, map : dict[int, str],  p : callable) -> str:
  n = int(s)
  r = ""
  for k, v in map.items():
    if p(n, k): r += v

  return r or s

print(FizzBuzz("15",
               {3: "Fizz", 5: "Buzz", 7: "Buff"},
               (lambda x, y: x % y == 0)))
```

```python
def plus_one(a): return a + 1

def add(a, b):
    total_sum = a
    for _ in range(b):
        total_sum = plus_one(total_sum)
    return total_sum

def multiply(a, b):
  c = 0
  for _ in range(b):
    c = add(c, a)
  return c
```

```python
import time

def tick_tock(n):
  for i in range(1, n + 1):
    print("tick..." if i % 2 else "tock...")
    time.sleep(1)

class My:
  def __bool__(self):
    r = int(time.time())
    return False if r % 2 else True

def is_empty(l):
  if l:
    print("Full")
  else:
    print("Empty")

is_empty(My())
is_empty(0)
is_empty("Hello")
is_empty([])
```

```python
def f(s : str) -> bool:
  tmp = set()
  for ch in s:
    if ch.isalpha():
      tmp.add(ch.upper())

  return len(tmp) == 26

print(f("The quick brown fox jumps over the yellow lazy dog."))
```
